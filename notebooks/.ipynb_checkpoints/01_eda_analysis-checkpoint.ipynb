## 3. Distribution Plots &mdash; Raw Features with Derived Cutoffs

Each histogram below marks the **percentile-derived threshold** the pipeline actually computed (printed in the pipeline log above), so the cutoff is visually traceable to a specific point in this dataset's own distribution.

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))

# --- days_inactive | admin ---
ax = axes[0, 0]
admin_inactive = users_df[users_df["privilege_level"] == "admin"]["days_inactive"]
p75_admin = np.percentile(admin_inactive, 75)
sns.histplot(admin_inactive, bins=15, ax=ax, color="#3b82f6")
ax.axvline(p75_admin, color="red", linestyle="--", label=f"P75 = {p75_admin:.1f} days")
ax.set_title("days_inactive | privilege_level == 'admin'")
ax.legend()

# --- days_inactive | power-user ---
ax = axes[0, 1]
pu_inactive = users_df[users_df["privilege_level"] == "power-user"]["days_inactive"]
p75_pu = np.percentile(pu_inactive, 75)
sns.histplot(pu_inactive, bins=15, ax=ax, color="#a78bfa")
ax.axvline(p75_pu, color="red", linestyle="--", label=f"P75 = {p75_pu:.1f} days")
ax.set_title("days_inactive | privilege_level == 'power-user'")
ax.legend()

# --- tenure_days (all users) ---
ax = axes[1, 0]
p10_tenure = np.percentile(users_df["tenure_days"], 10)
sns.histplot(users_df["tenure_days"], bins=20, ax=ax, color="#10b981")
ax.axvline(p10_tenure, color="red", linestyle="--", label=f"P10 = {p10_tenure:.1f} days (new hire cutoff)")
ax.set_title("tenure_days (all 300 users)")
ax.legend()

# --- system_count (all users) ---
ax = axes[1, 1]
sns.histplot(users_df["system_count"], bins=7, ax=ax, color="#f59e0b", discrete=True)
ax.set_title("system_count (all 300 users) &mdash; range 1-7")

plt.tight_layout()
plt.show()
```

## 4. Anomaly Score Distribution & Risk Level Cutoffs

```python
p70 = users_df["anomaly_score"].quantile(0.70)
p84 = users_df["anomaly_score"].quantile(0.84)
p95 = users_df["anomaly_score"].quantile(0.95)

print(f"P70 = {p70:.1f}  (MEDIUM cutoff)")
print(f"P84 = {p84:.1f}  (HIGH cutoff / ~top 16% per brief's stated user anomaly rate)")
print(f"P95 = {p95:.1f}  (CRITICAL cutoff, combined with privilege/sensitivity condition)")

fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(users_df["anomaly_score"], bins=30, ax=ax, color="#3b82f6")
for val, label, color in [(p70, "P70 (MEDIUM)", "#eab308"),
                           (p84, "P84 (HIGH)", "#f97316"),
                           (p95, "P95 (CRITICAL)", "#ef4444")]:
    ax.axvline(val, color=color, linestyle="--", label=f"{label} = {val:.1f}")
ax.set_title("anomaly_score distribution (300 users) with derived risk-level cutoffs")
ax.legend()
plt.show()
```

## 5. Label-Free Validation (Section 9 of Build Blueprint)

No ground-truth label files were provided. The validation strategy below replaces precision/recall with:

1. **Flagged-vs-population comparison** &mdash; do flagged accounts sit at the statistical extremes?
2. **Flagging-rate consistency** &mdash; does our overall flagging rate align with the brief's stated ~16% (users) / ~41% (events)?

```python
fig, ax = plt.subplots(figsize=(10, 5))
order = ["LOW", "MEDIUM", "HIGH", "CRITICAL"]
sns.boxplot(data=users_df, x="risk_level", y="anomaly_score", order=order, ax=ax,
            palette={"LOW": "#10b981", "MEDIUM": "#eab308", "HIGH": "#f97316", "CRITICAL": "#ef4444"})
ax.set_title("anomaly_score by risk_level &mdash; flagged accounts should sit at the extremes")
plt.show()
```

```python
total_users = len(users_df)
flagged_users = (users_df["risk_level"].isin(["HIGH", "CRITICAL"])).sum()
flagging_rate_users = flagged_users / total_users

total_events = len(events_df)
# event-level anomaly flag column name may vary; adjust if needed
if "event_is_anomaly" in events_df.columns:
    flagged_events = events_df["event_is_anomaly"].sum()
    flagging_rate_events = flagged_events / total_events
    print(f"Event flagging rate: {flagged_events}/{total_events} = {flagging_rate_events:.1%}  (brief states ~41%)")
else:
    print("Event-level anomaly flag column not found under expected name &mdash; check src/model.py output columns.")

print(f"User flagging rate (HIGH+CRITICAL): {flagged_users}/{total_users} = {flagging_rate_users:.1%}  (brief states ~16%)")
```

### Limitation Statement

Ground-truth label files (`identity_users_labels.csv`, `identity_events_labels.csv`) referenced in supplementary documentation were **not present** in the provided `sample_data` folder &mdash; confirmed via direct folder listing (only `identity_users.csv` and `identity_events.csv` exist).

Every threshold in this system (P75/P90/P10 cutoffs for stale accounts, new hires, contractors, exports, etc.) is derived from this dataset's own distribution at runtime and printed in the pipeline log above &mdash; not hardcoded. The Isolation Forest `contamination` parameters (0.16 for users, 0.41 for events) are taken directly from the anomaly rates *stated in the problem brief* as modeling hyperparameters, not as labels we validate against.

We report flagging-rate consistency with these stated rates as a sanity check (above), and show that flagged accounts (`HIGH`/`CRITICAL`) sit at the statistical extremes of the `anomaly_score` distribution (box plot above) relative to `LOW`/`MEDIUM` accounts.

We explicitly **cannot** report precision/recall/F1 against a hidden ground truth, because none was provided to us with this dataset.

## 7. Sparse Event History &amp; Confidence (Novelty 7.3)

Users with zero or very few events in the full year receive lower confidence scores regardless of their static risk profile, since less behavioural evidence supports the score.

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.scatterplot(data=users_df, x="events_per_user", y="confidence", hue="adjusted_risk_level",
                 palette={"LOW": "#10b981", "MEDIUM": "#eab308", "HIGH": "#f97316", "CRITICAL": "#ef4444"},
                 ax=ax, alpha=0.7)
ax.set_title("Confidence vs. events_per_user &mdash; sparse history -> lower confidence")
plt.show()

zero_event_users = users_df[users_df["events_per_user"] == 0]
print(f"{len(zero_event_users)} users with zero events in 365 days.")
print("Their confidence scores:")
print(zero_event_users["confidence"].describe())
```

## 9. Export Outputs (sample_report.json, figures, graph HTML)

Saves the required deliverables into `output/` in the project root.

```python
import json

flagged = users_df[users_df["adjusted_risk_level"].isin(["CRITICAL", "HIGH"])].copy()
print(f"Flagged accounts (CRITICAL+HIGH): {len(flagged)}")

reports = []
for uid, row in flagged.iterrows():
    user_events = events_df[events_df["user_id"] == uid]
    recurrence = detect_recurrence_pattern(user_events) if 'detect_recurrence_pattern' in dir() else {}

    payload = {
        "user_id": uid,
        "username": row["username"],
        "department": row["department"],
        "privilege_level": row["privilege_level"],
        "days_inactive": int(row["days_inactive"]),
        "system_count": int(row["system_count"]),
        "blast_radius": int(row.get("blast_radius", 0)),
        "anomaly_score": float(row["anomaly_score"]),
        "adjusted_score": float(row["adjusted_score"]),
        "risk_level": row["risk_level"],
        "adjusted_risk_level": row["adjusted_risk_level"],
        "exception_tags": row["exception_tags"],
        "is_stale_admin": int(row.get("is_stale_admin", 0)),
        "is_stale_power_user": int(row.get("is_stale_power_user", 0)),
        "sod_violations_count": int(row.get("sod_violations_count", 0)),
        "sod_violations": data_dict.get("sod_violations", {}).get(uid, []),
        "high_sensitivity_export_count": int(row.get("high_sensitivity_export_count", 0)),
        "admin_op_off_hours_count": int(row.get("admin_op_off_hours_count", 0)),
        "confidence": float(row["confidence"]),
        "confidence_basis": row["confidence_basis"],
        "recurrence_info": recurrence,
    }
    reports.append(generate_explanation(payload))

output_dir = os.path.join(project_root, "output")
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, "sample_report.json")

with open(output_path, "w") as f:
    json.dump(reports, f, indent=2, default=str)

print(f"Saved {len(reports)} reports to {output_path}")
```
